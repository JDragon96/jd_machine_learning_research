```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

sys.path.append("..")
import Distributions
import Statistic
```

```python
d1 = np.array([[1, 2], [3, 1], [5, 2], [3, 3]])
d2 = np.array([[6, 6], [8, 5], [10, 6], [8, 7]])
mu1 = np.mean(d1, axis=0)
sigma1 = Statistic.getCovMatrix(d1, 2)
mu2 = np.mean(d2, axis=0)
sigma2 = Statistic.getCovMatrix(d2, 2)
```

```python
def decision_1D_boundary(data, m1, m2, sig, w1, w2):
    data = np.array(data)
    sig1 = np.array(sig)
    
    if len(sig1) == len(data):
        data = data.T
    inv_sig = np.linalg.inv(sig1)
    a = inv_sig @ (m1 - m2).reshape(1, -1).T
    b = np.log(w1) - np.log(w2) - 0.5* m1 @ inv_sig @ m1.T + 0.5 * m2 @ inv_sig @ m2.T
    
    x1_arg = a[0][0]
    a /= x1_arg
    b /= x1_arg

    return -a[1][0], -b
    
def eq_1d(m, b, x):
    return m * x + b
```

```python
for i in range(10):
    dist1 = i / 10
    dist2 = (10 - i) / 10
    
    #      
    m, b = decision_1D_boundary(d1, mu1, mu2, sigma1, dist1, dist2)

    plt.title(f"w1 : {dist1} / w2 : {dist2}")
    plt.plot(d1[:, 0], d1[:, 1], '.', color='b')
    plt.plot(d2[:, 0], d2[:, 1], '.', color='r')
    plt.plot([0, 10], [eq_1d(m, b, 0), eq_1d(m, b, 10)], color='black')
    plt.show()
```
