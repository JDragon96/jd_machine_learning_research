```python
import numpy as np
import matplotlib.pyplot as plt
import copy
from scipy.stats import multivariate_normal
```

## 데이터 준비

```python
def gen_data(k=3, dim=2, points_per_cluster=200, lim=[-10, 10]):
    x = []
    mean = np.random.randint(-30, 30, size=k * dim).reshape(k, dim)
    for i in range(k):
        cov = np.random.rand(dim, dim*30)
        cov = np.matmul(cov, cov.T)
        _x = np.random.multivariate_normal(mean[i], cov, points_per_cluster)
        x += list(_x)
    x = np.array(x)
    
    if(dim == 2):
        fig = plt.figure()
        ax = fig.gca()
        ax.scatter(x[:,0], x[:,1], s=3, alpha=0.4)
        ax.autoscale(enable=True) 
    return x

def multivariable_gaussian(data, mu, sig):
    """
    >>> data: (N, dim)
    >>> mu  : (dim)
    >>> sig : (dim, dim)
    """
    new_data = copy.deepcopy(data)
    dim = np.shape(data)[1]
    sig_abs = sig[0][0] * sig[1][1] - sig[1][0] * sig[0][1]
    inv_sig = np.linalg.inv(sig)
    sig_tile = np.tile(inv_sig, (len(data), 1, 1))
    u = pow(2 * np.pi, dim / 2) * sig_abs
    
    e_total = (np.expand_dims(new_data, axis=1) - mu)
    result = np.exp(e_total @ sig_tile @ e_total.transpose(0, 2, 1) * -0.5) / u
    return result.reshape(len(data))
    
multivariable_gaussian([[1,2], [3,4], [5,6], [1, 2]] , [1, 2], [[1, 0], [0, 1]])
multivariate_normal.pdf([[1,2], [3,4], [5,6], [1, 2]] , [1, 2], [[1, 0], [0, 1]])
```

```python
k = 3
dim = 2
iterations = 100

data = gen_data(points_per_cluster=1000)
d_size = len(data)
est_n = np.zeros(k)
est_w = np.zeros((d_size, k))
est_mu = np.random.randint(-10, 10, size=6).reshape(k, dim)
est_sig = np.tile(np.eye(2), (k, 1, 1))
est_pi = np.ones(k) / k
print(f"데이터 쉐입 : {np.shape(data)}")
```

```python
for itertaion in range(iterations):
    # E STEP
    for cluster in range(k):
        est_w[:, cluster] = est_pi[cluster] * np.array(multivariable_gaussian(data, est_mu[cluster], est_sig[cluster]))
    
    for index in range(d_size):
        est_w[index] /= sum(est_w[index])
    
    # M STEP
    est_n = np.sum(est_w, axis=0)
    for cluster in range(k):
        est_mu[cluster] = est_w[:, cluster].T @ data / (est_n[cluster] + 1e-9)
        u = (np.expand_dims(data, axis=1) - est_mu[cluster])
        est_sig[cluster] = (u.transpose(0, 2, 1)  @ u).transpose(1, 2, 0) @ est_w[:, cluster] / (est_n[cluster] + 1e-9)
        est_pi[cluster] = est_n[cluster] / d_size
        
    print(est_mu)
    plt.plot(data[:, 0], data[:, 1], '.', c='black')
    plt.plot(est_mu[0, 0], est_mu[0, 1], '*', c='r')
    plt.plot(est_mu[1, 0], est_mu[1, 1], '*', c='r')
    plt.plot(est_mu[2, 0], est_mu[2, 1], '*', c='r')
    plt.show()
        
        
```
